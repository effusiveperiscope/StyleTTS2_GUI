```python
from gui_models import find_models
from core import StyleTTS2Core
import numpy as np
import json
import os

config = {
    'models_dir': 'Models'
}
model_specs = find_models(config)
LABELS_INDEX = "episodes_labels_index.json"
master_file_1 = "D:\MLP_Samples\AIData\Master file"
model_folder = 'Multi1_35_48k'
chars = ['Twilight', 'Rainbow', 'Pinkie', 'Fluttershy', 'Rarity',
    'Applejack', 'Celestia', 'Luna', 'Cadance', 'Apple Bloom', 'Sweetie Belle',
    'Scootaloo', 'Starlight', 'Trixie', 'Spike', 'Big Mac', 'Zecora',
    'Cozy Glow', 'Cheerilee', 'Shining Armor', 'Discord', 'Diamond Tiara',
    'Gilda', 'Chrysalis', 'Tirek', 'Flim', 'Flam']

def longpath(path):
    import platform
    path = os.path.abspath(path)
    if 'Windows' in platform.system() and not path.startswith('\\\\?\\'):
        path = u'\\\\?\\'+path.replace('/','\\')
        return path
    else:
        return path

with open(LABELS_INDEX, encoding='utf-8') as f:
    j = json.load(f)

core = StyleTTS2Core(device='cuda')
char_vecs = {}

key = model_folder
spec = model_specs[key]
# Assume that the style index has already been created if it can be found
if spec['style_index'] is not None:
    print(f"{key} style index already exists")
    exit()
print(f"Processing model {key}")
print(spec['config'])
print(spec['ckpt'])
core.load_model(spec['config'], spec['ckpt'])

for k,v in j.items():
    print(f"Processing {k}")
    for l in v['lines']:
        orig_file_path = longpath(
            l['orig_file'].replace('MASTER_FILE_1', master_file_1))
        char = l['parse']['char']
        noise = l['parse']['noise']
        
        if char not in chars:
            continue
        if len(noise): # only allow clean audios
            continue

        t = core.style_from_path_with_components(orig_file_path)
        if t is None:
            continue
        style_vec, ref_s, ref_p = t
        if not char in char_vecs:
            char_vecs[char] = []

        char_vecs[char].append(style_vec.cpu().numpy())

for char in chars:
    if char not in char_vecs:
        continue
    a = np.array(char_vecs[char])
    print(f"Calculating median for {char}")
    med = np.median(a, axis=0)
    char_vecs[char] = med.tolist()

with open(os.path.join(config['models_dir'], model_folder, 'styles.json'), 'w') as f:
    json.dump(char_vecs, f)
```
